```python
import pandas as pd
data = {
    "timestamp": [
        "2025-06-27 18:01:02", "2025-06-27 18:01:10", "2025-06-27 18:01:15",
        "2025-06-27 18:02:01", "2025-06-27 18:02:03", "2025-06-27 18:03:12",
        "2025-06-27 18:05:22", "2025-06-27 18:05:25", "2025-06-27 18:06:14"
    ],
    "attacker_ip": [
        "192.168.1.10", "192.168.1.11", "192.168.1.10",
        "192.168.1.12", "192.168.1.10", "192.168.1.11",
        "192.168.1.13", "192.168.1.11", "192.168.1.14"
    ],
    "username_attempted": [
        "root", "admin", "root",
        "admin", "guest", "admin",
        "user", "admin", "admin"
    ],
    "password_attempted": [
        "123456", "password", "admin",
        "admin123", "123456", "root",
        "user", "admin", "admin"
    ],
    "port": [22, 22, 22, 22, 22, 22, 22, 22, 22]
}

df = pd.DataFrame(data)
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Show first few rows
df.head()
```

```python
import matplotlib.pyplot as plt

# Count attack attempts per IP
ip_counts = df['attacker_ip'].value_counts()

# Plot bar chart
plt.figure(figsize=(8,5))
ip_counts.plot(kind='bar', color='orange')
plt.title('Top Attacker IPs')
plt.xlabel('Attacker IP')
plt.ylabel('Number of Attempts')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Count most common usernames attackers tried
username_counts = df['username_attempted'].value_counts()

# Plot bar chart of top usernames
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
username_counts.plot(kind='bar', color='purple')
plt.title('Top Usernames Attempted by Attackers')
plt.xlabel('Username')
plt.ylabel('Number of Attempts')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
##  MITRE ATT&CK Mapping

- **Tactic**: Credential Access
- **Technique**: Brute Force (SSH)
- **MITRE ID**: T1110.001
- **Why**: The logs show repeated login attempts on SSH port 22 using common usernames like `root`, `admin`, and `guest`, which matches brute-force behavior.
```
